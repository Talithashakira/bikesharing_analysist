**2.2. Cek Data Ordinal**

Memastikan bahwa nilai untuk kolom season, mnth, hr, weekday, weathersit, dan yr berada pada rentang nilai yang sesuai dengan informasi dataset



```python
ordinal_checks = {
    "season": [1, 2, 3, 4],
    "mnth": list(range(1, 13)),  # 1 - 12
    "hr": list(range(0, 24)),  # 0 - 23 
    "weekday": list(range(0, 7)),  # 0 - 6
    "weathersit": [1, 2, 3, 4],  # 1 - 4
    "yr": [0, 1]  # 0 = 2011, 1 = 2012
}

invalid_found = False

for col, valid_values in ordinal_checks.items():
    if col in hour_df.columns:
        invalid_values = hour_df[~hour_df[col].isin(valid_values)]
        if not invalid_values.empty:
            print(f"Invalid values found in '{col}' (hour.csv):")
            display(invalid_values)
            invalid_found = True
    
    if col in day_df.columns:
        invalid_values = day_df[~day_df[col].isin(valid_values)]
        if not invalid_values.empty:
            print(f"Invalid values found in '{col}' (day.csv):")
            display(invalid_values)
            invalid_found = True  

if not invalid_found:
    print("Tidak ada data yang invalid di kolom ordinal.")
```

**2.3. Cek Data Nominal**

```python
# Cek apakah holiday dan workingday hanya berisi 0 atau 1
nominal_cols = ["holiday", "workingday"]

invalid_found = False  

for col in nominal_cols:
    if col in hour_df.columns:
        invalid_values = hour_df[~hour_df[col].isin([0, 1])]
        if not invalid_values.empty: 
            print(f"Invalid values found in '{col}' (hour.csv):")
            display(invalid_values)
            invalid_found = True  

    if col in day_df.columns:
        invalid_values = day_df[~day_df[col].isin([0, 1])]
        if not invalid_values.empty:  
            print(f"Invalid values found in '{col}' (day.csv):")
            display(invalid_values)
            invalid_found = True  

# Jika tidak ada data yang invalid
if not invalid_found:
    print("Tidak ada data yang invalid di kolom 'holiday' dan 'workingday'.")
```

### 3. Cleaning Data

**3.1 Merubah Tipe data dteday**

Kita ingin mengubah kolom dteday dari object (string) menjadi datetime agar lebih mudah dalam analisis waktu.

```python
# Ubah tipe data dteday menjadi datetime
hour_df["dteday"] = pd.to_datetime(hour_df["dteday"])
day_df["dteday"] = pd.to_datetime(day_df["dteday"])

print("Tipe data setelah konversi:")
print("Hour Data:", hour_df.dtypes["dteday"])
print("Day Data:", day_df.dtypes["dteday"])
```

**3.2. Mengganti Kolom Hour menjadi Time Periode**

```python
def categorize_time(hour):
    if 5 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 16:
        return 'Afternoon'
    elif 17 <= hour <= 20:
        return 'Evening'
    else:
        return 'Night'

hour_df['time_period'] = hour_df['hr'].apply(categorize_time)

time_period_mapping = {'Morning': 1, 'Afternoon': 2, 'Evening': 3, 'Night': 4}
hour_df['time_period'] = hour_df['time_period'].map(time_period_mapping)

categorical_cols = ['season', 'mnth', 'weekday', 'weathersit', 'yr', 'holiday', 'workingday']
numerical_cols = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']

time_period_df = hour_df.groupby(['dteday', 'time_period'], observed=True).agg({
    **{col: 'mean' for col in numerical_cols},  # Mean untuk numerik
    **{col: lambda x: x.mode()[0] for col in categorical_cols}  # Mode untuk kategorikal
}).reset_index()

time_period_df = time_period_df.sort_values(by=['dteday', 'time_period']).reset_index(drop=True)

time_period_df.head(4)
```

### Pertanyaan 1:
Bagaimana perbedaan rata-rata jumlah peminjaman sepeda antara pengguna casual dan registered setiap bulan selama tahun 2011 dan 2012??

**Langkah Analisis**
1. Filter data untuk tahun 2011 (yr == 0) dan 2012 (yr == 1)
2. Kelompokkan berdasarkan bulan (mnth)
3. Hitung rata-rata jumlah peminjaman untuk casual dan registered
4. Visualisasikan dengan grafik garis untuk membandingkan tren bulanan

```python
# Filter data untuk tahun 2012
df_2012 = day_df[day_df["yr"] == 1]

# Hitung rata-rata peminjaman per bulan untuk casual dan registered
monthly_avg = df_2012.groupby("mnth")[["casual", "registered"]].mean().reset_index()

# Visualisasi data
plt.figure(figsize=(10, 5))
sns.lineplot(x="mnth", y="casual", data=monthly_avg, label="Casual", marker="o")
sns.lineplot(x="mnth", y="registered", data=monthly_avg, label="Registered", marker="o")

plt.title("Rata-rata Peminjaman Sepeda per Bulan (2012)")
plt.xlabel("Bulan")
plt.ylabel("Rata-rata Jumlah Peminjaman")
plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "Mei", "Jun", "Jul", "Agu", "Sep", "Okt", "Nov", "Des"])
plt.legend()
plt.grid(True)
plt.show()
```

**Insight Rata-Rata Peminjaman sepeda (Tahun 2011 dan 2012)**
- Pengguna registered secara konsisten memiliki jumlah peminjaman yang jauh lebih tinggi dibandingkan pengguna casual di kedua tahun
- Kedua jenis pengguna menunjukkan pola musiman yang mirip dengan puncak di bulan-bulan musim panas (Juni-September)
- Terjadi peningkatan signifikan pada jumlah peminjaman dari 2011 ke 2012 untuk kedua kelompok pengguna

### Pertanyaan 3:
Pada jam berapa rata-rata jumlah peminjaman sepeda harian tertinggi dan terendah dalam periode Januari - Desember 2012, berdasarkan data historis?

**Langkah Analisis**
...

```python
# Filter data untuk tahun 2012
df_2012_hour = hour_df[hour_df["yr"] == 1]

# Hitung rata-rata peminjaman per jam
hourly_avg = df_2012_hour.groupby("hr")["cnt"].mean().reset_index()

# Temukan jam dengan peminjaman tertinggi dan terendah
peak_hour = hourly_avg.loc[hourly_avg["cnt"].idxmax()]
lowest_hour = hourly_avg.loc[hourly_avg["cnt"].idxmin()]

# Visualisasi data
plt.figure(figsize=(10, 5))
sns.lineplot(x="hr", y="cnt", data=hourly_avg, marker="o", color="b")

plt.axvline(x=peak_hour["hr"], color="r", linestyle="--", label=f"Peak: {int(peak_hour['hr'])}:00")
plt.axvline(x=lowest_hour["hr"], color="g", linestyle="--", label=f"Lowest: {int(lowest_hour['hr'])}:00")

plt.title("Rata-rata Peminjaman Sepeda per Jam (2012)")
plt.xlabel("Jam")
plt.ylabel("Rata-rata Jumlah Peminjaman")
plt.xticks(range(0, 24))
plt.legend()
plt.grid(True)
plt.show()

print(f"🚴‍♂️ Jam dengan peminjaman tertinggi: {int(peak_hour['hr'])}:00 dengan rata-rata {peak_hour['cnt']:.2f} peminjaman")
print(f"🚴‍♂️ Jam dengan peminjaman terendah: {int(lowest_hour['hr'])}:00 dengan rata-rata {lowest_hour['cnt']:.2f} peminjaman")
```

**Insight dari Visualisasi Rata-Rata Peminjaman Sepeda per Jam:**
- **Jam Puncak (Peak):** **Pukul 17:00** – Jumlah peminjaman tertinggi.  
- **Jam Terendah (Lowest):** **Pukul 04:00** – Jumlah peminjaman terendah.  
 
- **Dua Puncak Utama**  
  **Pagi (07:00 - 09:00):** Lonjakan signifikan, kemungkinan karena orang pergi bekerja/sekolah.  
   **Sore (16:00 - 18:00):** Peminjaman tertinggi terjadi sekitar pukul 17:00, kemungkinan karena orang pulang kerja/sekolah.  

- **Jam Sepi (Dini Hari - Subuh)**  
  **00:00 - 05:00:** Aktivitas peminjaman sangat rendah, hampir nol pada pukul 04:00.  

- **Aktivitas Stabil di Siang Hari**  
   **10:00 - 15:00:** Peminjaman relatif stabil, kemungkinan digunakan untuk keperluan santai atau rekreasi.
